```python
import pandas as pd
data = pd.read_csv ('robot3bac.csv')
data.head()
```

```python
#x=data.iloc[:,[2,3]]
#y=data.iloc[:,[0,1]]
x_data=data.drop(['tt1','tt2','tt3'],axis=1)
y_data=data[['tt1','tt2','tt3']]
```

```python
x_data
```

```python
y_data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=7)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping

model=Sequential()
model.add(Dense(128, activation='relu', input_shape = (3,))) 
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3,  activation='linear'))
model.summary()
```

```python
model.compile(loss='mse',metrics=['mean_absolute_error'])
history= model.fit(x_train, y_train, batch_size=100, epochs=500, verbose=1, validation_split=0.2, callbacks=[EarlyStopping(monitor='val_loss', patience=80)])
```

```python
score=model.evaluate(x_test,y_test,verbose=0)
print('Test loss',score[0])
print('Test accuracy',score[1])
```

```python
import numpy as np
predictions=model.predict(x_test)
y_test=np.array(y_test)
print("Gia tri du doan:", predictions[3])
print("Gia tri thuc:",y_test[3])
```
